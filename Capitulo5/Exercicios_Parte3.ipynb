```python
import pandas as pd
import numpy as np

# Carregar dataset
paises = pd.read_csv('paises.csv', delimiter=';')

# Mostrar colunas
print(paises.columns)
```

1. Carregue o Dataset paises.csv e em seguida mostre: 

    a.Quais são os países da OCEANIA; 

    b.Quantos países são da OCEANIA; 
    
Dica: para busca de padrões textuais no Pandas, use métodos da subclasse str da Series. Ex: series.str.contains(‘texto’)

```python
# Separar países da Oceania
oceania = paises[paises['Region'].str.contains('OCEANIA')]

# Mostrar quais são os países da Oceania
print('Os países da Oceania são:')
print(oceania['Country'].values)

# Mostrar quantos países são da Oceania
print(f"São {len(oceania)} países da Oceania.")
```

2. Encontre o nome e a região do país que possui a maior população segundo este Dataset; 

```python
# Separar país com a maior população
maiorPopulacao = paises.nlargest(1, 'Population')

# Mostrar nome e região desse país
print(f"O país com maior população é {maiorPopulacao['Country'].item()}, que está na região {maiorPopulacao['Region'].item()}.")
```

3. Agrupe os países por Regiões. Em seguida, mostre a média de alfabetização (Literacy (%)) de cada região do planeta;

```python
# Agrupar países por regiões
regioes = paises.groupby('Region')

# Calcular e mostrar média de alfabetização de cada região
qtdPaises = regioes.count()['Country'] #qtd paises por região
somaAlf = regioes.sum()['Literacy (%)'] #soma da alfabetização
mediaAlf = somaAlf/qtdPaises

print(mediaAlf)
```

4.  Busque o nome de todos os países do Dataset que não possuem costa marítima (Coastline (coast/area ratio) == 0) e guarde-os em um novo arquivo chamado noCoast.csv; 

```python
# Separar países que não possuem costa marítima
semCostaMaritima = paises['Country'][paises['Coastline (coast/area ratio)'] == 0]

# Salvar em um arquivo
semCostaMaritima.to_csv('noCoast.csv', sep=';')
```

5. Faça uma função que receba a taxa de mortalidade de cada país (Deathrate) e retorne o texto ‘Balanced’ caso o valor seja < 9 e ‘Urgent’ caso contrário. Em seguida, crie um campo no Dataset chamado ‘Humanitarian Help’ que receba estes valores para cada país. No final, mostre o Dataset para verificar se a inserção da nova coluna foi feita com sucesso.

```python
# Criar função para verificação do nível da taxa de mortalidade
def humanitarianHelp(x):
    if x<9:
        return 'Balanced'
    else:
        return 'Urgent'
    
# Criar nova coluna no dataset com dados dessa verificação
paises['Humanitarian Help'] = paises['Deathrate'].apply(humanitarianHelp)

# Mostrar 10 primeiros itens do dataset
paises.head(10)
```
