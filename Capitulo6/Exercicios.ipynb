```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Importar datasets
paises = pd.read_csv('paises.csv', delimiter=';')
space = pd.read_csv('space.csv', delimiter=';')

# Mostrar colunas dos datasets
print('As colunas do dataset paises.csv são: ', paises.columns)
print('As colunas do dataset space.csv são: ', space.columns)
```

1. Por meio do dataset paises.csv, trace dois gráficos de linhas em um mesmo plano cartesiano, um mostrando a taxa de mortalidade (Deathrate) e outro a taxa de natalidade (Birthrate) dos países da América do Norte;

```python
# Separar países da América do Norte
americaNorte = paises[paises['Region'].str.contains('NORTHERN AMERICA')]

# Separar eixos
x = americaNorte['Country']
y_mortalidade = americaNorte['Deathrate']
y_natalidade = americaNorte['Birthrate']

# Adicionar as informações do gráfico
plt.title('Taxas de Natalidade e Mortalidade da América do Norte')
plt.xlabel('Países')
plt.ylabel('Taxas (%)')

# Plotar gráfico
plt.plot(x, y_natalidade, 'o-b', label='Taxa de Natalidade')
plt.plot(x, y_mortalidade, 'o-r', label='Taxa de Mortalidade')
plt.legend()
```

2. Por meio do dataset space.csv, trace um gráfico em barras mostrando quantas empresas espaciais diferentes os EUA e a CHINA possuem; Dica: não se esqueça de retirar os resultados repetidos

```python
# Separar diferentes empresas dos EUA e China
empresas_eua = space[space['Location'].str.contains('USA')]['Company Name'].unique()
empresas_china = space[space['Location'].str.contains('China')]['Company Name'].unique()

# Separar eixos
eixo_x = ['EUA', 'China']
eixo_y = [len(empresas_eua), len(empresas_china)]

# Adicionar informações ao gráfico
plt.title('Quantidade de empresas espaciais diferentes')
plt.xlabel('Países')

# Plotar gráfico
plt.bar(eixo_x, eixo_y, color='orange')
```

3. Por meio do dataset space.csv, trace um gráfico em torta ilustrando a porcentagem de missões da empresa Roscosmos que deram certo e que deram errado;

```python
# Separar missões da empresa Roscosmos
roscosmos = space[space['Company Name'].str.contains('Roscosmos')]

# Separar missões que deram certo e errado
roscosmos_certo = roscosmos[roscosmos['Status Mission'].str.contains('Success')]
roscosmos_errado = roscosmos[roscosmos['Status Mission'].str.contains('Failure')]

# Criar e plotar gráfico
plt.title('Status das Missões da Empresa Roscosmos')
plt.pie(x=[len(roscosmos_certo), len(roscosmos_errado)],
        labels=['% Missão que deram certo', '% Missão que deram errado'],
        autopct='%1.1f%%')
```
